# 2. 핵심 개념 이해
**상관 분석은 변수들이 서로 독립적인지 아니면 서로 영향을 주고받는지를 알아내는 분석 방법이다.**<br/>
**파이썬의 seaborn 라이브러리 패키지에서 제공하는 데이터셋으로 데이터과학 방법론에 따라 상관 분석 프로젝트를 수행할 수 있다.**<br/>

- **상관분석**<br/>
  **상관분석(correlation analysis)은 두 변수가 어떤 선형적 관계에 있는지를 분석하는 방법이다.**<br/>
  **두 변수는 서로 독립적이거나 상관된 관계일 수 있는데, 두 변수의 관계의 강도를 상관관계라고 한다.**<br/>
  **상관 분석에서는 상관관계의 정도를 나타내는 단위로 모상관 계수 p를 사용한다.**<br/>
  **인과 관계를 분석하는 회귀 분석과 달리 상관 계수(correlation coefficient)는 두 변수가 연관된 정도를 나타낼 뿐 인과 관계를 설명하지 않으므로 정확한 예측치를 계산할 수는 없다**<br/><br/>
  **상관 분석의 유형은 다음과 같다.**<br/>
  - 단순 상관 분석(simple correlation analysis) : 두 변수가 어느 정도 강한 관계에 있는지 측정한다.
  - 다중 상관 분석(multiple correlation analysis) : 세 개 이상의 변수 간 관계의 강도를 측정한다. 이때 다른 변수와의 관계를 고정하고 두 변수 간 관계의 강도를 타나내는 것을 편상관분석(partial correlation analysis)라고 한다.<br/><br/>

- **상관계수p**<br/>
  **상관계수p는 변수 간 관계의 정도(0~1)와 방향(+, -)을 하나의수치로 요약해주는 지수로 -1에서 +1 사이의 값을 가진다.**<br/>
  **상관 계수가 + 이면 양의 상관관계이며 한 변수가 증가하면 다른 변수도 증가한다. 상관 계수가 - 이면 음의 상관관계이며 한 변수가 증가할 때 다른 변수는 감소한다.**<br/>
  - 0.0 ~ 0.2 : 상관관계가 거의 없다.
  - 0.2 ~ 0.4 : 약한 상관관계가 있다.
  - 0.4 ~ 0.6 : 상관관계가 있다.
  - 0.6 ~ 0.8 : 강한 상관관계가 있다.
  - 0.8 ~ 1.0 : 매우 강한 상관관계가 있다.
<br/><br/><br/>
- **피어슨 상관 계수**<br/>
  **상관 계수 중에서 많이 사용하는 것은 피어슨 상관 계수(pearson correlation coefficient) 또는 Pearson's r 이다. 피어슨 상관 계수는 r로 표현한다.**<br/>
- **상관 분석 결과의 시각화**<br/>
  **상관 분석 결과를 시각화할 때는 두 변수의 관계를 보여주는 산점도나 히트맵을 많이 사용한다.**

```python
# 프로젝트에 사용할 타이타닉호 생존자에 대한 데이터는 파이썬의 seaborn 라이브러리 패키지에서 제공하는 타이타닉 데이터를 사용한다.
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")
titanic.to_csv("C:/Users/kgw19/Data-science-project/titanic/titanic.csv", index = False)
```

```python
# age와 embarked, deck, embark_town 항목 중에 결측값(누락된 값)이 있다. 결측값이 있으면 정확한 분석을 할 수 없으므로
# 치환해서 채워야 한다. age의 결측값은 중앙값으로 치환하고 embarked와 embark_town은 최빈값으로 치환한다.
# deck은 형식이 category 이므로 최빈 category로 바꾸어 채워넣는다.

titanic.isnull().sum() # 타이타닉 데이터에 결측값이 있는지 확인. 실행 결과를 보면 age, embarked, deck, embark_town에서 결측값 발견
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'].value_counts()
# embarked 열의 결측값을 최빈값으로 치환
# embarked 열에서 최빈값이 "S"임을 확인
# embarked 열에서 결측값을 "S"로 치환
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')  
titanic['embark_town'].value_counts()                  
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southamton')  # embark_town 열에서 최빈값이 "Southampton'임을 확인
titanic['deck'].value_counts()                                        # embark_town 열에서 결측값을 "Southampton"으로 치환
```

```python
titanic['deck'] = titanic['deck'].fillna('C')
titanic.isnull().sum()
```

```python
titanic.info() # 타이타닉 데이터의 기본 정보 확인
```

```python
titanic.survived.value_counts()
```

```python
# 차트를 그려 데이터를 시각적으로 탐색하기

import matplotlib.pyplot as plt
f, ax = plt.subplots(1, 2, figsize = (10, 5))
titanic['survived'][titanic['sex'] == 'male'].value_counts().plot.pie(explode = [0, 0.1], autopct = "%1.1f%%", ax = ax[0], shadow = True) # 남성 승객의 생존률
titanic['survived'][titanic['sex'] == 'female'].value_counts().plot.pie(explode = [0, 0.1], autopct = "%1.1f%%", ax = ax[1], shadow = True) # 여성 승객의 생존률
ax[0].set_title('Survived(Male)')
ax[1].set_title('Survived(Female)')
plt.show()
```

# 3. 데이터 모델링
**타이타닉호 승객의 속성과 생존 사이에 어떤 상관관계가 있는지 분석하는 모델을 만들기**<br/>
**상관분석은 pandas 패키지의 데이터프레임에서 corr()함수를 사용하고, 상관 계수는 피어슨 상관계수를 사용한다.**<br/>
**상관계수는 연속형 데이터에서만 구할 수 있으므로 데이터 형식이 int64, boolean인 속성 8개에 대해서만 상관 분석을 진행한다.**

```python
titanic['survived'] = titanic['survived'].astype('int64')
titanic.info()
```

```python
numeric_columns = titanic.select_dtypes(include=['int64', 'float64', 'bool'])  # 상관계수 분석을 위해 연속형 데이터인 float64, int64, boolean 데이터를 가진 행만 선택
titanic_corr = numeric_columns.corr(method = 'pearson') # 피어슨 상관계수 사용
titanic_corr

# 아래의 결과를 확인해보면 남성 성인은 생존과 음의 상관관계를 보여준다. 
# 객실등급은 음의 상관관계
# 객실요금은 양의 상관관계
# 동행없이 혼자 탑승한 경우에도 생존율이 떨어진다는 상관관계도 볼 수 있다.
```

```python
# 변수 사이의 상관 계수를 히트맵으로 시각화하기
'''
변수 중 age를 히트맵에 나타내려면 카테고리 형식이어야 하므로 10살 단위로 0부터 7까지 카테고리 갚으로 치환한다. 그리고 상관 계수를 계산하려면
숫자 데이터이어야 하므로 sex 변수값을 0과 1로 치환한다. 우리가 가지고 있는 데이터에는 동행한 형제의 수 sibsp와 부모/자식의 수 parch가 있는데
이를 더해서 가족의 수를 새로운 변수 family로 추가하여 히트맵에 표현
'''
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
titanic['age2'] = titanic['age'].apply(category_age)
titanic['sex'] = titanic['sex'].map({'male' : 1, 'female' : 0})
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
titanic.to_csv("C:/Users/kgw19/Data-science-project/titanic/titanic3.csv", index = False)
```

```python
heatmap_data = titanic[['survived', 'sex', 'age2', 'family', 'pclass', 'fare']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0, square = True, cmap = colormap, linecolor = 'white', annot = True, annot_kws = {'size':10})
plt.show()
```
